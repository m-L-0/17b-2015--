```python
_author__ = 'freedom'
import tensorflow as tf
import numpy as np

category ={
    0:"T-shirt/top（T恤）",
    1:"Trouser（裤子）",
    2:"Pullover（套衫）",
    3:"Dress（裙子）",
    4:"Coat（外套）",
    5:"Sandal（凉鞋）",
    6:"Shirt（汗衫）",
    7:"Sneaker（运动鞋）",
    8:"Bag（包）",
    9:"Ankle boot（踝靴）"
}

def loadMNIST():
    from tensorflow.examples.tutorials.mnist import input_data
    mnist = input_data.read_data_sets('/home/haiyang/tensorflow/data/',one_hot=True)
    return mnist
def KNN(mnist):
    train_x,train_y = mnist.train.next_batch(5000)
    test_x,test_y = mnist.train.next_batch(200)

    xtr = tf.placeholder(tf.float32,[None,784])
    xte = tf.placeholder(tf.float32,[784])
    distance = tf.sqrt(tf.reduce_sum(tf.pow(tf.add(xtr,tf.negative(xte)),2),reduction_indices=1))

    pred = tf.argmin(distance,0)

    init = tf.initialize_all_variables()

    sess = tf.Session()
    sess.run(init)

    right = 0
    for i in range(200):
        ansIndex = sess.run(pred,{xtr:train_x,xte:test_x[i,:]})
        if np.argmax(test_y[i]) == np.argmax(train_y[ansIndex]):
            right += 1.0
            print ('预测值是：',category[np.argmax(train_y[ansIndex])])
            print ('真实值是：',category[np.argmax(test_y[i])])
        else: 
            print ('预测值是：*',category[np.argmax(train_y[ansIndex])])
            print ('真实值是：*',category[np.argmax(test_y[i])])
    print("正确预测：",right)
    print("错误预测：",200-right)
    accracy = right/200.0
    print ("正确率：",accracy)

if __name__ == "__main__":
    mnist = loadMNIST()
    KNN(mnist)
```
